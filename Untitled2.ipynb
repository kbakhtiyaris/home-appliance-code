```python
import pandas as pd, numpy as np, os, glob
from pathlib import Path

data_dir = out_dir # Use the output directory from the previous cell
csv_files = sorted(glob.glob(str(data_dir / "*.csv")))

csv_file = max(csv_files, key=lambda p: os.path.getsize(p))
print("Selected CSV file:", csv_file)

# reading first few rows to detect timestamp column
first5 = pd.read_csv(csv_file, nrows=5)
print("Columns:", list(first5.columns))
display(first5)

# Common timestamp column candidates
candidates = [c for c in first5.columns if 'time' in c.lower() or 'date' in c.lower() or 'timestamp' in c.lower()]
print("Timestamp candidates detected:", candidates)
```

```python
# Task 1: Dataset characterization
from IPython.display import display, Markdown
import pandas as pd
import numpy as np

# Load the dataset and combine Date and Time into a single datetime index
df = pd.read_csv(csv_file)
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df = df.set_index('DateTime').drop(['Date', 'Time'], axis=1)


print("Dataframe loaded. Shape:", df.shape)
display(df.head())

cols = df.columns.tolist()
n_vars = len(cols)
start, end = df.index.min(), df.index.max()
time_span = end - start

# Attempt to infer sampling frequency using median difference
dt = df.index.to_series().diff().dropna().dt.total_seconds()
median_dt = dt.median() if len(dt)>0 else np.nan
cv = dt.std()/dt.mean() if len(dt)>0 and dt.mean()!=0 else np.nan

# Data types summary
dtypes = df.dtypes.apply(lambda x: x.name).to_dict()

display(Markdown(f"**Number of variables (columns):** {n_vars}"))
display(Markdown(f"**Columns:** {cols}"))
display(Markdown(f"**Time span:** {start} to {end}  (span = {time_span})"))
display(Markdown(f"**Median sampling interval (seconds):** {median_dt}"))
display(Markdown(f"**Coefficient of variation of intervals (CV):** {cv:.3f}"))
display(Markdown(f"**Inferred sampling regularity:** {'Regular' if (cv<0.1) else 'Irregular' if (cv>0.5) else 'Mostly regular / slightly variable' }"))
display(Markdown(f"**Data types (per column):**"))
display(pd.DataFrame.from_dict(dtypes, orient='index', columns=['dtype']))
```

```python
# Task 2: Four V's
import sys
from IPython.display import display, Markdown
import numpy as np

# Volume
n_rows, n_cols = df.shape
memory = df.memory_usage(deep=True).sum()
rows_per_day = None
if n_rows>1:
    total_seconds = (df.index.max() - df.index.min()).total_seconds()
    rows_per_day = n_rows / (total_seconds/86400) if total_seconds>0 else np.nan

display(Markdown(f"**Volume**\n- Total records: {n_rows}\n- Columns: {n_cols}\n- Memory usage (bytes): {memory}\n- Approx rows/day: {rows_per_day:.2f}"))

# Velocity: inter-arrival times distribution
dt = df.index.to_series().diff().dropna().dt.total_seconds()
display(Markdown(f"**Velocity**\n- Mean interval (s): {dt.mean():.3f}\n- Median interval (s): {dt.median():.3f}\n- Std interval (s): {dt.std():.3f}\n- CV: {(dt.std()/dt.mean() if dt.mean()!=0 else np.nan):.3f}"))

# Variety
num_numeric = df.select_dtypes(include=[np.number]).shape[1]
num_non_numeric = n_cols - num_numeric
display(Markdown(f"**Variety**\n- Numeric columns: {num_numeric}\n- Non-numeric columns: {num_non_numeric}\n- Different sensor-like variables: {n_cols}"))

# Veracity note (full analysis in Tasks 3-5)
display(Markdown("**Veracity**: See Tasks 3–5 (Latency, Completeness, Accuracy) for data quality analysis."))

# Plot histogram of inter-arrival times
import matplotlib.pyplot as plt
plt.figure(figsize=(6,3))
plt.hist(dt.values, bins=60)
plt.xlabel('Inter-arrival time (seconds)')
plt.title('Histogram of inter-arrival times')
plt.tight_layout()
plt.show()
```

```python
# Task 3: Latency Analysis
from IPython.display import display, Markdown
import numpy as np

dt = df.index.to_series().diff().dropna().dt.total_seconds()
mean_dt = dt.mean()
std_dt = dt.std()
cv = std_dt/mean_dt if mean_dt!=0 else np.nan

# Gaps larger than 2x median interval
median_dt = dt.median()
gaps = dt[dt > 2*median_dt]
n_gaps = gaps.shape[0]
gap_summary = gaps.describe()

display(Markdown(f"- Mean interval (s): {mean_dt:.3f}\n- Std interval (s): {std_dt:.3f}\n- CV: {cv:.3f}"))
display(Markdown(f"- Median interval (s): {median_dt:.3f}\n- Number of gaps > 2×median: {n_gaps}"))

# Classify latency according to lecture thresholds (approx)
if mean_dt < 0.01:
    latency_class = "Ultra-low (<10ms)"
elif mean_dt < 0.1:
    latency_class = "Low (10–100ms)"
elif mean_dt < 1:
    latency_class = "Medium (100ms–1s)"
elif mean_dt < 10:
    latency_class = "High (1–10s)"
else:
    latency_class = "Batch (>10s)"
display(Markdown(f"**Latency class:** {latency_class}"))

# Visualize timeline and mark gaps
import matplotlib.pyplot as plt
plt.figure(figsize=(10,2))
y = np.zeros(len(df.index))
plt.plot(df.index, y, '|', markersize=10)
for g in gaps.index:
    plt.axvline(g, color='red', alpha=0.3)
plt.title('Event timeline (gaps > 2×median in red)')
plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
# Task 4: Completeness
from IPython.display import display, Markdown
import numpy as np

# If series is roughly regular, infer expected freq
median_dt = df.index.to_series().diff().dropna().dt.total_seconds().median()

display(Markdown("Could not infer a single regular frequency; computing simple missingness metrics."))
display(Markdown(f"- Total records: {df.shape[0]}"))
miss_pct = df.isna().sum() / df.shape[0] * 100
display(Markdown("**Missing % per column:**"))
display(miss_pct.sort_values(ascending=False))

# Optional: visual missingness using matplotlib (simple heatmap)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plt.imshow(df.isna().T, aspect='auto', interpolation='nearest')
plt.xlabel('Time index (sample number)')
plt.yticks(range(len(df.columns)), df.columns)
plt.title('Missingness heatmap (columns vs. samples)')
plt.colorbar(label='isna')
plt.tight_layout()
plt.show()
```

```python
# Task 5: Accuracy checks
from IPython.display import display, Markdown
import numpy as np
import matplotlib.pyplot as plt

numeric = df.select_dtypes(include=[np.number])
display(Markdown(f"Numeric columns detected: {list(numeric.columns)}"))

# Range validation: teacher-style generic plausible ranges for energy sensors
range_checks = {}
for c in numeric.columns:
    vals = numeric[c].dropna()
    if vals.empty:
        continue
    rng = (vals.min(), vals.max())
    q1, q3 = np.percentile(vals, [25,75])
    iqr = q3 - q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    outliers = vals[(vals < lower) | (vals > upper)]
    range_checks[c] = {'min': rng[0], 'max': rng[1], 'n_outliers': outliers.shape[0], 'lower_bound': lower, 'upper_bound': upper}

display(pd.DataFrame(range_checks).T)

# Show boxplot for numeric columns (teacher-style)
plt.figure(figsize=(8,4))
numeric.boxplot(rot=45)
plt.title('Boxplot of numeric columns (outliers shown)')
plt.tight_layout()
plt.show()

# Noise: variance and rolling mean comparison for first numeric column
if numeric.shape[1]>0:
    col = numeric.columns[0]
    vals = numeric[col].dropna()
    display(Markdown(f"**Column for rolling analysis:** {col}"))
    display(Markdown(f"- Variance: {vals.var():.6f}\n- Std: {vals.std():.6f}"))
    roll = vals.rolling(window=5, min_periods=1).mean()
    plt.figure(figsize=(10,3))
    plt.plot(vals.index, vals, label='Raw', alpha=0.6)
    plt.plot(roll.index, roll, label='Rolling mean (w=5)', lw=1.5)
    plt.legend()
    plt.title(f'Raw vs Rolling mean for {col}')
    plt.tight_layout()
    plt.show()
else:
    display(Markdown("No numeric columns found for accuracy checks."))

# Drift detection: rolling mean over larger window
if numeric.shape[1]>0:
    col = numeric.columns[0]
    roll_long = numeric[col].rolling(window=min(100, len(numeric))).mean()
    plt.figure(figsize=(10,3))
    plt.plot(numeric[col].index, numeric[col], alpha=0.3)
    plt.plot(roll_long.index, roll_long, lw=2)
    plt.title(f'Long-window rolling mean (drift check) for {col}')
    plt.tight_layout()
    plt.show()
```
